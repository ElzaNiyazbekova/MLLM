# Q5: Visualization – What Influences Tips? (25 marks)


```python
plt.figure(figsize=(14, 4))

sns.scatterplot(x='tip', y='total_bill',
alpha=0.7, data=df)
plt.title('Tip vs Total Bill')
plt.show()
```

```python
plt.figure(figsize=(14, 4))

df.groupby('day')['tip'].mean().plot(kind='bar')
plt.title('Average Tip by Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.show()
```

```python
plt.figure(figsize=(14, 4))
sns.histplot(df['tip'], kde=True)
plt.title('Tip distribution')
```

# Answer Q5.2: Interpret each plot in 1–2 sentences.


<p>In the first plot, we can see that the highest amount of tips is received on Sunday, then Saturday, which means that tips are mostly given on weekends.</p>

<p>In the second plot, we can see that people most often give tips between 2$ and 4$, while tips between $4 and $10 are less common.</p>
